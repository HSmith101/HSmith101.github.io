```python
def numbirred(n,p): #The number of irred. polys of degree n in F_p[x].
    L = []
    for i in range(1,n+1):
        if ZZ(i).divides(ZZ(n)):
            L.append(p^i*moebius(n/i))
        else:
            pass
    return 1/n*sum(L)
```

```python
def ordGSp(g,n): #This is the order of GSp_{2g}(n).
    primeorders = []
    h = factor(n)
    for i in range(0,len(h)):
        l = h[i][0]
        e = h[i][1]
        primeorders.append((l-1)*l^(g^2)*product(l^(2*index)-1,index,1,g)*(l^(2*g^2+g+1))^(e-1)) #I've used product to try to make things easy to change as I increase my dimension g.
    return product(primeorders)
```

```python
GSpOrders = [0,0]
for n in range(2,1001):
    GSpOrders.append(ordGSp(2,ZZ(n))) #g=2.
```

```python
def Frobmatrix2(a1,a2,p): #This is for the minimal order Z[pi,v].
    frob = Matrix([[0,0,p,-p*a1],[1,0,0,-a2],[0,0,0,-p],[0,1,0,-a1]])
    return frob
```

```python
def WeilPolys(p):
    list = []
    R.<x> = QQ[]
    for a3 in range(-ceil(4*sqrt(p)),floor(4*sqrt(p))+1): #This is from the first condition.
        for a2 in range(ceil(2*sqrt(p)*abs(a3)-2*p),floor(a3^2/4+2*p)+1): #This is from the second condition.
            f= R(x^4+a3*x^3+a2*x^2+p*a3*x+p^2)
            if f.is_irreducible():
                list.append([a3,a2])
                #print(f)
    return list
```

```python
def Mono2(bound,a3,a2,p):
    #bound is the arbitrary upper bound.
    print(a3,a2)
    nonmono = [] #This is to get my list on non-monogenic stuff easily.
    Frob = Frobmatrix2(a3,a2,p)
    #print Frob
    print("****************")
    #print latex(Frob)
    if ZZ(a2).valuation(p) == 0:
        prank = 2
    elif ZZ(a3).valuation(p) == 0:
        prank = 1
    else:
        prank = 0
    print("p-rank is %i." % prank)
    for n in range(2,bound):
        #print n
        try:
            if gcd(Frob.determinant(),n)!=1:
                #print "n is %i and the determinant is not 1." % n
                pass
            else:
                GSpsize = GSpOrders[n]
                #print "GSpsize is %i." % GSpsize
                R = Integers(ZZ(n)) #This works better than the quotient ring I was using.
                V = GL(2*2,R) #I will need to change this for different g.
                F = V(Frob)
                Fj = F
                j=1
                while GSpsize/numbirred(j,p)>j:
                    if not Fj.is_one():
                        #print "j is %i." % j
                        Fj = Fj*F
                        j = j+1
                    else:
                        ord = j
                        print("***********************************")
                        print("n is %i." % n)
                        print("The order of the Frobenius is %i." % ord)
                        print(factor(ord))
                        print("We have %i polys of the proper degree in Z/%iZ[x]." % (numbirred(ord,p),p))
                        split = GSpsize/ord
                        print("The number of primes %i splits into has to be %s." % (p,split))
                        nonmono.append(str(n))
                        break
                pass
        except:
            pass
    print(nonmono)
    return "%i & %i & %i & %s \\\\ \\hline" %(a3,a2,prank,", ".join(nonmono)) #join concatenates and the quotes tell it what to put between the elts of the list.
```
